```python
import seaborn as sns

df = sns.load_dataset('iris')
```

Draw scatter plot

```python
sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"],
).set(title = 'Kevin: Sepal length vs Sepal width',
     xlabel = 'Sepal length',
     ylabel = 'Sepal width')
```

```python
sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
).set(title = 'Kevin: Petal length vs Petal width',
     xlabel = 'Petal length',
     ylabel = 'Petal width')
```

```python
sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
            scatter_kws = {"color": "red"} # For color
).set(title = 'Kevin: Petal length vs Petal width',
     xlabel = 'Petal length',
     ylabel = 'Petal width')
```

```python
sns.set_style('dark')

sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"],
).set(title = 'Kevin: Sepal length vs Sepal width',
     xlabel = 'Sepal length',
     ylabel = 'Sepal width')
```

```python
sns.set_style('darkgrid')  # (darkgrid, whitegrid, dark, white, ticks)

sns.regplot(x = df["petal_length"],
            y = df["petal_width"],
            scatter_kws = {"color": "red"} # For color
).set(title = 'Kevin: Petal length vs Petal width, Dark Grid Style',
     xlabel = 'Petal length',
     ylabel = 'Petal width')
```

```python
sns.set_style('ticks')

sns.regplot(x = df["sepal_length"],
            y = df["sepal_width"],
).set(title = 'Kevin: Sepal length vs Sepal width',
     xlabel = 'Sepal length',
     ylabel = 'Sepal width')
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
import pandas as pd

gapminder = pd.read_csv(
    'https://raw.githubusercontent.com/holtzy/The-Python-Graph-Gallery/master/static/data/gapminderData.csv'
)
```

```python
gapminder.info()
```

```python
gapminder.head()
```

```python
gapminder.tail()
```

To use data 2007 above

```python
gapminder.continent.unique()
gapminder.year.unique()  

data = gapminder.loc[gapminder.year == 2007]
```

```python
data.shape
```

```python
sns.set_theme(style = 'darkgrid') 
sns.scatterplot(
    data = data,
    x = 'gdpPercap',
    y = 'lifeExp', 
    size = 'pop',
    hue = 'country',
    legend = False,
    sizes=(20, 2000) 
).set(title = 'Kevin: GDP per Capita vs Life expectancy', 
      xlabel = 'GDP per Capita',
      ylabel = 'Life expectancy')

plt.show
```

```python
data = gapminder.loc[(gapminder.year == 2007) & (gapminder.continent == 'Europe')]
```

```python
sns.set_theme(style = 'darkgrid') 
sns.scatterplot(
    data = data,
    x = 'gdpPercap',
    y = 'lifeExp', 
    size = 'pop',
    hue = 'country',
    legend = False,
    sizes=(20, 2000) 
).set(title = 'Kevin: GDP per Capita vs Life expectancy, Europe', 
      xlabel = 'GDP per Capita',
      ylabel = 'Life expectancy')

plt.show
```
